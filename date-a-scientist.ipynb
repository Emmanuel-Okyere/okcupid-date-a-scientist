```python
columns_to_drop = ['essay0','essay1', 'essay2', 'essay3', 'essay4', 'essay5', 'essay6', 'essay7','essay8', 'essay9']
newProfile = profiles.drop(columns_to_drop, axis = 1)
newProfile.head()
```

```python
import numpy as np
newProfile["body_type"].fillna("other", inplace = True)
newProfile["diet"].fillna("other", inplace = True)
newProfile["drinks"].fillna("other", inplace = True)
newProfile["drugs"].fillna("never", inplace = True)
newProfile["education"].fillna("other", inplace = True)
newProfile["height"].fillna(np.mean(newProfile["height"]), inplace = True)
newProfile["job"].fillna("other", inplace = True)
newProfile["offspring"].fillna("other", inplace = True)
newProfile["pets"].fillna("other", inplace = True)
newProfile["religion"].fillna("other", inplace = True)
newProfile["smokes"].fillna("no", inplace = True)
newProfile["ethnicity"].fillna("other", inplace = True)
newProfile["sign"].fillna("other", inplace = True)
newProfile["speaks"].fillna("other", inplace = True)
newProfile["income"].replace(-1, np.mean(newProfile["income"]), inplace = True)
newProfile["height"] = newProfile["height"].apply(lambda x: np.mean(newProfile["height"]) if x < 30 else x)
newProfile["norm_income"] = newProfile["income"].apply(lambda x: (x-np.mean(newProfile["income"]))/np.std(newProfile["income"]) if x > 1 else 0)
newProfile["sign_clean"] = newProfile["sign"].str.split().str.get(0)
newProfile.info()
```
